# Importação do Dataset ENADE 2017

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
name_path_csv = "DADOS_ENADE_2018/xaa"
arquivo = open(name_path_csv)
df = pd.read_csv(arquivo, delimiter=";")
df['NT_GER'] = [str(val).replace(',', '.') for val in df['NT_GER']]
#cria a coluna Notas convertendo para numerico
df['Notas']=pd.to_numeric(df.NT_GER,errors='coerce')
```

# Tentando identificar padões que possibilitam uma nota boa

```python
#Quantidade de alunos por categoria, em determinada variável 

var_media = 'NT_GER' #variavel que vou fazer a media

def qtd_alunos_por_categoria(var_media,var_analisada):
    alunos = df[(df[var_media] >= 0) & (df[var_analisada] != '.') & (df[var_analisada] != '*')]
    alunos[var_analisada].value_counts().plot.pie(title=var_analisada)
    
    alunos[var_analisada].value_counts()
```

```python
df = df[df['TP_PRES'] == 555] #presente na prova


def notas_medias_na_variavel(var_analisada):
    groups = []
    result_groups = []
    opcoes = []
    for index in df[var_analisada]:
        if (index not in opcoes) & (index != '*') & (index != '.'):
            opcoes.append(index)            
    opcoes.sort()
    print(opcoes)
    opcoes
    for i in range(0,len(opcoes)):
        grp = df[df[var_analisada] == opcoes[i]]
        result_array = grp['NT_GER']
        groups.append(grp)
        result_groups.append(result_array)
        print(result_array)
        

    plt.boxplot(result_groups,patch_artist=True, labels=opcoes)
    plt.title(var_analisada)
    plt.ylabel('Mean grades')
    #plt.xlabel('family income')
    #plt.savefig('DADOS_ENADE_2017/QE_I'+str(aux)+'.png', format='png')
    plt.show()
```
